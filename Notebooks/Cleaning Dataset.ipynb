```python
data = pd.read_csv("../Data/nba_2020_shooting.csv")
```

```python
data.columns
```

```python
data
```

```python
#In the NBA there is many players that just play a few minutes during the year beacause they join the league only for a few games because of injuries mainly. We really don´t need those players. They are usually going to have some NULL values due to the fact that they haven´t even shoot a free throw or a triple. 
cleaning = data.dropna(thresh=25, axis=0)
```

```python
cleaning
```

```python
#Then we are going to drop the players that have played less than 10 games, since there statistics are not reliable. 
cleaning2 = cleaning.drop(cleaning[cleaning["G"] < 10].index)
```

```python
cleaning2
```

```python
cleaning2.columns
```

```python
#Since we only wnat shooting percentages there are many columns we don´t need.
cleaning3 = cleaning2.drop(["Dist.", "%FGA Dunks", "3P Proportion Astd", "# Dunks", "3P% Corner 3s", "MP", "# Heaves", "Att. Heaves", "%3PA Corner 3s", "2P Proportion Astd", "16-3P FG%", "10-16 FG%", "3-10 FG%", "0-3 FG%", "3P Proportion", "16-3P Proportion", "10-16 Proportion", "3-10 Proportion", "0-3 Proportion", "2P Proportion"], axis=1)
```

```python
cleaning3
```

```python
#Let´s reset the index
cleaning4 = cleaning3.reset_index()
```

```python
cleaning4
```

```python
#Finally let´s eliminate the old index
cleaning5 = cleaning4.drop(["index"], axis =1)
```

```python
cleaning5.sample(25)
```

```python
final_data = cleaning5
```

```python
final_data.to_csv(r"final_data.csv")
```
